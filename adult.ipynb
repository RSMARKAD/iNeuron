```python
data['relationship'].value_counts().plot(kind='bar')
plt.title('relationship of family members')
plt.xlabel('Family Members')
plt.ylabel('Counts')
plt.show()
```

```python
data['education-num'].value_counts().sort_index().plot(kind='bar', color = ['#354259'])
data['education'].value_counts().sort_index().plot(kind='bar',color = ['#F73D93'])
plt.title('education and education-num for Family')
plt.xlabel('Education and num')
plt.ylabel('Counts')
plt.show()
```

```python
data['marital-status'].value_counts().sort_index().plot(kind='bar', color = 'g')
data['occupation'].value_counts().sort_index().plot(kind='bar',color = ['#D2691E'])
plt.title('marital-status and occupation for Family')
plt.xlabel('marital-status and occupation')
plt.ylabel('Counts')
plt.show()
```
